```python
df = df.rename(columns = {'Product_Category':'Product Category','Marital_Status':'Marital Status',
                     'Cust_name':'Customer Name', 'Product_ID':'Product ID', 'User_ID':'User ID' })
```

```python
df.columns
```

```python
df.head(1)
```

```python
ax = sns.countplot(x='Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df1 = df.groupby(['Gender'],as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = True)
```

```python
df1
```

```python
ax1 = sns.barplot(x ='Gender', y = 'Amount', data = df1)
for bars in ax1.containers:
    ax1.bar_label(bars)
```

```python
sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
```

```python
df2 = df.groupby(['State'],as_index= False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)
df2.plot(kind = 'bar', x= 'State', y = 'Orders', figsize=(10,6), color = 'skyblue')
```

```python
df3 = df.groupby(['State'],as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)
colors = ['red', 'blue', 'green', 'purple', 'orange', 'cyan', 'magenta', 'yellow', 'brown', 'pink']
df2.plot(kind = 'bar', x= 'State', y = 'Orders', figsize=(10,6), color = colors, legend = False)
```
